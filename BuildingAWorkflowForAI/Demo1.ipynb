```python
import yfinance as yf
import pandas as pd
```

```python
# Define the start and end dates
start_date = '2024-04-01'
end_date = '2024-05-01'

# Define the list of tickers
tickers = ['AAPL', 'MSFT']

historical_prices = yf.download(tickers, start=start_date, end=end_date)
```

```python
historical_prices.head()
```

```python
sp500_tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()
# Filter out tickers that end with '.B'
sp500_tickers = [ticker for ticker in sp500_tickers if '.B' not in ticker]
```

```python
sp500_tickers
```

```python
start_date = '2000-01-01'
end_date = '2024-05-01'
historical_prices = yf.download(sp500_tickers, start=start_date, end=end_date, auto_adjust=False)
```

```python
historical_prices = historical_prices.loc[:, historical_prices.columns.get_level_values(0) == 'Adj Close']
historical_prices.columns = historical_prices.columns.droplevel(0)
```

```python
historical_prices.head()
```

```python
MIN_REQUIRED_NUM_OBS_PER_TICKER = 100

# Count the number of observations for each ticker
ticker_counts = historical_prices.count()
ticker_counts
```

```python
# Filter tickers with at least MIN_REQUIRED_NUM_OBS_PER_TICKER observations
valid_tickers_msak = ticker_counts[ticker_counts >= MIN_REQUIRED_NUM_OBS_PER_TICKER].index
historical_prices = historical_prices[valid_tickers_msak]
```
