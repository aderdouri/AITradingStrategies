```python
import numpy as np
import yfinance as yf
from utils import plot_returns
```

```python
# Download S&P 500 front month futures data
prices = yf.download(['ES=F', 'ZN=F'], auto_adjust=False)["Adj Close"].resample("ME").last()
# Calculate logarithmic returns
log_returns = np.log(prices).diff()
# Set portfolio weights
# YOUR CODE HERE
weights = np.array([.6, .4])
# Calculate weighted returns
# YOUR CODE HERE
weighted_returns = log_returns.mul(weights, axis=1)
# YOUR CODE HERE
# Calculate portfolio returns
portfolio_returns = weighted_returns.sum(axis=1)
```

```python
plot_returns(portfolio_returns)
```
